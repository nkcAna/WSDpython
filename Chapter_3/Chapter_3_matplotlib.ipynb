## Matplotlib library

**Matplotlib** is an important library for visualizing data in Python, and it offers a wide array of plotting methods that can accommodate both numerical and categorical variables. To provide a comprehensive overview, we will start by creating a simulated dataset that includes both types of variables (numerical, categorical). Then, we will discuss essential components of matplotlib such as **importing** the library, creating **subplots**, and **saving** figures. 

### Creating a Simulated Dataset

To create a dataset that can demonstrate a variety of plots, we will use both numerical and categorical variables. For this, we will use the **numpy** and **pandas** libraries to create a DataFrame.


```python
import numpy as np
import pandas as pd

# Number of data points
n = 100

# Generate numerical data
weight = np.random.normal(150, 40, n)
height = np.random.normal(60, 10, n)
food_intake = np.random.normal(5, 2, n)

# Generate categorical data
animal_types = ['Cow', 'Sheep', 'Pig', 'Goat']
health_status = ['Healthy', 'Sick']

animal_type_data = np.random.choice(animal_types, n)
health_status_data = np.random.choice(health_status, n)

# Create DataFrame
df = pd.DataFrame({
    'Weight': weight,
    'Height': height,
    'Food_Intake': food_intake,
    'Animal_Type': animal_type_data,
    'Health_Status': health_status_data
})
```

### Importing Matplotlib

The library can be imported using the import statement.

```python
# import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
import matplotlib.pyplot as plt

# Create the figure and axes
fig, ax = plt.subplots()

# Create a color map for animal types
colors = {'Cow': 'red', 'Sheep': 'green', 'Pig': 'blue', 'Goat': 'purple'}

# Scatter plot for Weight vs. Height, colored by Animal_Type
for animal_type, color in colors.items():
    subset = df[df['Animal_Type'] == animal_type]
    ax.scatter(subset['Weight'], subset['Height'], c=color, label=animal_type, alpha=0.6)

# Title
ax.set_title('Weight vs Height by Animal Type')

# Legend
ax.legend(title='Animal Type')

# X-axis and Y-axis labels
ax.set_xlabel('Weight (kg)')
ax.set_ylabel('Height (cm)')

# Major ticks for x-axis and y-axis
ax.set_xticks(np.arange(50, 250, 20))
ax.set_yticks(np.arange(40, 80, 5))

# Spines: making top and right spines invisible
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Save the figure
fig.savefig('weight_vs_height_by_animal_type.png')

# Show the plot
plt.show()
```

```python
# Simulated dataset 
import pandas as pd
import numpy as np

np.random.seed(0)
n = 50
data = {
    'AnimalType': np.random.choice(['Cow', 'Sheep', 'Chicken'], n),
    'Weight': np.random.uniform(1, 500, n),
    'Height': np.random.uniform(1, 200, n)
}

df_farm = pd.DataFrame(data)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate numerical data
x_numeric = np.linspace(0, 10, 100)
y_numeric = np.sin(x_numeric)

# Generate categorical data
categories = ['A', 'B', 'C', 'D']
y_categorical = np.random.choice(categories, 100)

# Create DataFrame
df = pd.DataFrame({'X': x_numeric, 'Y': y_numeric, 'Category': y_categorical})

# Create the figure and axes
fig, ax = plt.subplots()

# Plot data using data from DataFrame df
ax.plot(df['X'], df['Y'], color='blue', marker='o', label='Sin curve')

# Title
ax.set_title('Anatomy of a Matplotlib Plot')

# Legend
ax.legend()

# Major ticks and labels for x-axis
ax.set_xticks(np.arange(0, 11, 1))
ax.set_xticklabels(np.arange(0, 11, 1))

# Major ticks and labels for y-axis
ax.set_yticks(np.linspace(-1, 1, 11))
ax.set_yticklabels(np.round(np.linspace(-1, 1, 11), 2))

# Y-axis label
ax.set_ylabel('Y-values')

# X-axis label
ax.set_xlabel('X-values')

# Spines: making top and right spines invisible
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Save the figure
fig.savefig('anatomy_plot_using_df.png')

# Show the figure
plt.show()
```

```python
# Line Plot

plt.plot(df_farm.index, df_farm['Weight'])
plt.xlabel('Index')
plt.ylabel('Weight')
plt.title('Line Plot of Animal Weights')
plt.show()
```

```python
# Bar Plot

plt.bar(df_farm['Weight'], df_farm['Height'])
plt.show()
```

```python
# Titles and Labels

plt.plot(df_farm.index, df_farm['Height'])
plt.title('Line Plot')
plt.xlabel('Index')
plt.ylabel('Height')
plt.show()
```

```python
# Multiple Plots
plt.subplot(1, 2, 1)
plt.plot(df_farm.index, df_farm['Height'])

plt.subplot(1, 2, 2)
plt.bar(df_farm['Weight'], df_farm['Height'])
plt.show()
```
